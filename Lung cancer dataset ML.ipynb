```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
```

```python
# Encode categorical variables
encoder = LabelEncoder()
data['GENDER'] = encoder.fit_transform(data['GENDER'])
data['LUNG_CANCER'] = encoder.fit_transform(data['LUNG_CANCER'])
```

```python
# Split the data into features (X) and target (y)
X = data.drop(columns=['LUNG_CANCER'])
y = data['LUNG_CANCER']
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize models
models = {
    'SVM': SVC(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Linear Regression': LinearRegression()  # Linear Regression is not ideal for classification but included as requested
}
```

```python
# Train and evaluate models
results = {}
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    # Predict on the test set
    y_pred = model.predict(X_test)
    # Evaluate performance
    if name == 'Linear Regression':
        mse = mean_squared_error(y_test, y_pred)
        results[name] = {'Mean Squared Error': mse}
    else:
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)
        results[name] = {'Accuracy': accuracy, 'Classification Report': report}

results
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Initialize models
new_models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'AdaBoost': AdaBoostClassifier(random_state=42)
}

# Train and evaluate new models
new_results = {}
for name, model in new_models.items():
    # Train the model
    model.fit(X_train, y_train)
    # Predict on the test set
    y_pred = model.predict(X_test)
    # Evaluate performance
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, zero_division=0)
    new_results[name] = {'Accuracy': accuracy, 'Classification Report': report}

new_results
```
